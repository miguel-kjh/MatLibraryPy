```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plotData(dfBoost: pd.DataFrame, dfSwig: pd.DataFrame, title: str = "", colum: str = ""):
    plt.plot(dfBoost[colum], label = 'Boost')
    plt.plot(dfSwig[colum],  label = 'Swig')
    plt.title(title)
    plt.xlabel("Size")
    plt.ylabel("Times")
    plt.legend()
```

```python
dfBoost = pd.read_csv('tiempos_boost.csv')
dfSwig  = pd.read_csv('tiempos_swig.csv')

dfBoost.info()
```

```python
plotData(dfBoost, dfSwig, title = "Matrix Multiplication", colum = 'Mult')
```

```python
plotData(dfBoost, dfSwig, title = "Matrix Sum", colum = 'Sum')
```

```python
plotData(dfBoost, dfSwig, title = "Reverse operation", colum = 'Inv')
```

```python
plotData(dfBoost, dfSwig, title = "Save operation", colum = 'Save')
```
